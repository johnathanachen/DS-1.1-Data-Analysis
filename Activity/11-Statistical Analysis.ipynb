# Histogram

```python
x_1 = np.random.normal(-5, 1, 3000) # mp.random.normal(loc=0.0, scale=1.0, size=None)
x_2 = np.random.normal(2, 3, 7000)
x = np.concatenate((x_1, x_2))
```

```python
np.histogram(x)
```

```python
plt.hist(x, bins=20, density=1)
plt.show()
```

# Gaussian Mixture Model (GMM)

```python
gmm = mixture.GaussianMixture(n_components=2)
gmm.fit(x.reshape(-1,1))
print("means: ", gmm.means_)
print("covariances: ", gmm.covariances_)
print("weights: ", gmm.weights_)
# equal to mp.random.normal(loc=means, scale=covariances, size=weights)
```

```python
z = gmm.sample(10000)
plt.hist(z[0], bins=20, density=1)
plt.show()
```
